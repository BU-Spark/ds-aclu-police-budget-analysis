```python
earnings21 = pd.DataFrame(pd.read_csv('./Data/employees-2021.csv', encoding='latin1'))


earnings21['TOTAL_GROSS']=earnings21['TOTAL_GROSS'].apply(lambda x: float(x.replace(",","")))

fpath = Path(mpl.get_data_path(), "fonts/ttf/GT-America-Standard-Bold.ttf")
```

Data Visualizations
	-Pie/Graph of Boston City payroll BPD vs non BPD <br/>

	-Pie/Graph of Boston City payroll BPD vs non BPD 

```python
bpd_earnings21 = earnings21[earnings21['DEPARTMENT_NAME']=='Boston Police Department']
nonbpd_earnings21 = earnings21[earnings21['DEPARTMENT_NAME']!='Boston Police Department']
colors = ["#FCAA17", "#FCAA17", "#EF404E",'#552564','#0055AA',
    '#130F54','#3C3532','#691B40','#231F20','#EFECEA',
    '#FFE06A','#FFE06A','#FABEAF','#A3DBE3','#CCCCCC' ]
font = fpath


# make data
x = ['BPD', 'Non BPD']
y = [bpd_earnings21['TOTAL_GROSS'].sum(),
     nonbpd_earnings21['TOTAL_GROSS'].sum()]
# plot
fig, ax = plt.subplots()

fig.set_figheight(10)
fig.set_figwidth(10)


cont = ax.bar(x,y, color=['#EF404E',"#EFECEA"])
ax.set(ylim=(0, 1600000000),yticks=range(0,1600000000,300000000))
plt.ticklabel_format(style='plain', axis='y', useLocale=True)
plt.xlabel('Department', fontsize=16, font=font)
plt.ylabel('Amount Spent on Payroll', fontsize=16, font=font)
plt.title('City of Boston Payroll Breakdown',fontsize=18, font=font)
plt.grid(axis='y')


locale.setlocale(locale.LC_ALL, '')
labels = []
for i in y:
    labels.append(locale.currency(i, grouping=True))
    
ax.bar_label(container=cont, labels=labels,fontsize=14,font=font)
plt.xticks(font=fpath)
plt.yticks(font=fpath)



plt.show()
```

```python
bpd_earnings21 = earnings21[earnings21['DEPARTMENT_NAME']=='Boston Police Department']
nonbpd_earnings21 = earnings21[earnings21['DEPARTMENT_NAME']!='Boston Police Department']
colors = ["#FCAA17", "#FCAA17", "#EF404E",'#552564','#0055AA',
    '#130F54','#3C3532','#691B40','#231F20','#EFECEA',
    '#FFE06A','#FFE06A','#FABEAF','#A3DBE3','#CCCCCC' ]
font = fpath


# make data
x = ['Non BPD','BPD']
y = [ nonbpd_earnings21['TOTAL_GROSS'].sum() 
     / len(nonbpd_earnings21['TOTAL_GROSS'].index),
     bpd_earnings21['TOTAL_GROSS'].sum() 
     / len(bpd_earnings21['TOTAL_GROSS'].index)]
# plot
fig, ax = plt.subplots()

fig.set_figheight(10)
fig.set_figwidth(10)


cont = ax.bar(x,y, color=['#EF404E',"#EFECEA"])
ax.set(ylim=(0, 150000),yticks=range(0,150000,30000))
plt.ticklabel_format(style='plain', axis='y', useLocale=True)
plt.title('On average, BPD employees\' salaries are nearly double the amount of non-BPD city employees',fontsize=18, font=font)


locale.setlocale(locale.LC_ALL, '')
labels = []
for i in y:
    labels.append(locale.currency(i, grouping=True))


fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.bar_label(container=cont, labels=labels,fontsize=14,font=font)
plt.xticks(font=fpath)
plt.yticks(font=fpath)
#ax.yaxis.set_major_formatter('${x:1.0f}')



plt.show()
```

```python
bpd_earnings21 = earnings21[earnings21['DEPARTMENT_NAME']=='Boston Police Department']
nonbpd_earnings21 = earnings21[earnings21['DEPARTMENT_NAME']!='Boston Police Department']

    
colors = ['#EFECEA','#FFE06A','#FABEAF','#A3DBE3','#CCCCCC']
font = fpath

# make the data
x = []
y = []

ls = []
for i in earnings21['DEPARTMENT_NAME'].unique():
    ls.append(((earnings21[earnings21['DEPARTMENT_NAME']==i]['TOTAL_GROSS'].sum()),i))

ls.sort(reverse=True)
#print(ls)
for i,j in ls:
    y.append(i)
    x.append(j)



#y.sort(reverse=True)



# plot
fig, ax = plt.subplots()

fig.set_figheight(10)
fig.set_figwidth(10)

colors = []
for y2 in y[0:20]:
    if y2 > np.array(y[0:20]).mean()+(np.array(y[0:20]).mean()/2):
        colors.append('#EF404E')
    else:
        colors.append('#CCCCCC')
cont = ax.barh(x[0:20],y[0:20], color=colors)
ax.set(xlim=(0, 550000000),xticks=range(0,550000000,100000000))
plt.ticklabel_format(style='plain', axis='x', useLocale=True)
plt.yticks(rotation = 0) # Rotates X-Axis Ticks by 45-degrees
plt.ylabel('Department', fontsize=16, font=font)
plt.xlabel('Amount Spent on Payroll', fontsize=16, font=font)
plt.title('City of Boston Payroll Breakdown, Top 20 Largest Payrolls',fontsize=18, font=font)
plt.grid(axis='x')


locale.setlocale(locale.LC_ALL, '')
labels = []
for i in y[0:20]:
    labels.append(locale.currency(i, grouping=True))
    
ax.bar_label(container=cont, labels=labels,fontsize=14,font=font, rotation=0)
plt.xticks(font=fpath)
plt.yticks(font=fpath)


plt.show()
```
