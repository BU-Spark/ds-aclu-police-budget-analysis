### **********Male vs Female***********

```python
#FY22
```

```python
import pandas as pd

# Read in the data as a Pandas DataFrame
df = pd.read_csv('unit_assignment_22.csv')
df.head(20)

```

```python
# Count the number of males and females
num_males = df[df['Sex'] == 'M'].shape[0]
num_females = df[df['Sex'] == 'F'].shape[0]

# Display the results
print(f'Number of males: {num_males}')
print(f'Number of females: {num_females}')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties

labels = ['BLACK', 'WHITE']
sizes = [num_males, num_females]
colors = ['#FCAA17', '#0055AA']

gt_america_reg = FontProperties(fname='GT-America-Standard-Regular.ttf')
gt_america_bold = FontProperties(fname='GT-America-Standard-Bold.ttf')

# Plot
fig1, ax1 = plt.subplots()
wedges, _, autotexts = ax1.pie(sizes, colors=colors, autopct='', startangle=90, textprops={'fontsize': 10})
ax1.axis('equal')

# Create a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig1.gca().add_artist(centre_circle)

# Add labels and percentage outside the chart with lines (without pointers)
for i, (p, autotext) in enumerate(zip(wedges, autotexts)):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    
    # Draw a line
    ax1.plot([x, 1.2 * x], [y, 1.2 * y], '-',color='Black', linewidth=1.5)

    # Annotate the percentage
    percentage = f"{sizes[i] / sum(sizes) * 100:.1f}%"
    ax1.text(1.25 * x, 1.25 * y, f"{percentage}", fontsize=10, horizontalalignment=horizontalalignment,fontproperties=gt_america_reg)

# Set title
plt.title('Male vs Female Distribution in Police Data FY 22', 
          fontweight="bold", fontsize=12,loc='left',pad=30,x=0.2,fontproperties=gt_america_bold)

# Add legend
plt.legend(wedges, labels, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10)
plt.savefig("mvf22",bbox_inches = 'tight')
# Show the chart
plt.show()
```
