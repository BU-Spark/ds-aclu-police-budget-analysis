```python
import matplotlib.pyplot as plt
import numpy as np

# Data
departments = ['Boston Police Department', 'Boston Fire Department', 'Public Works Department', 'Property Management', 'BPS Facility Management', 'Parks Department'] 
budget_values = [78265402.51, 42902465.65, 4908683.73, 2416436.55, 2210072.78, 2090795.28]


gt_america_reg = FontProperties(fname='GT-America-Standard-Regular.ttf')
gt_america_bold = FontProperties(fname='GT-America-Standard-Bold.ttf')

# Sort data in ascending order
sorted_data = sorted(zip(departments, budget_values), key=lambda x: x[1])
sorted_departments, sorted_budget_values = zip(*sorted_data)

# Create horizontal bar graph
fig, ax = plt.subplots(figsize=(12, 6))  # Adjust the figure size
y_pos = np.arange(len(sorted_departments))
colors = ['#CCCCCC'] * (len(sorted_departments) - 1) + ['#EF404E']
ax.barh(y_pos, sorted_budget_values, align='center', color=colors)
ax.set_yticks(y_pos,fontsize=11,fontproperties=gt_america_bold)
ax.set_yticklabels(sorted_departments,fontsize=11,fontproperties=gt_america_reg)
ax.invert_yaxis()  # labels read top-to-bottom



plt.title('The total overtime pay of Boston Police Department ranked Highest in FY22,\n1.8 times that of BFD', 
          fontweight="bold", fontsize=18,loc='left',pad=10,x=-0.2,fontproperties=gt_america_bold)

# Change y-axis limits and labels
ax.set_ylim([-1, len(sorted_departments)])

# Add values to the bars
for i, v in enumerate(sorted_budget_values):
    ax.text(v + 1e6, i, '${:,.2f}'.format(v), color='#231F20', va='center',fontsize=11,fontproperties=gt_america_reg)

ax.set_xticks([])

# Remove box around the graph
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.tick_params(axis='both', which='both', bottom=False, top=False,left=False, right=False)
plt.savefig('od_boston_overtime_budget.png', dpi=300, bbox_inches='tight')
plt.show()
```
