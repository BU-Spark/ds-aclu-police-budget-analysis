```python
def convtostr(row):
    try:
        curzip = str(row['zip'])
        if len(curzip) == 0:
            return '00000'
        elif len(curzip) == 1:
            return '0000' + curzip
        elif len(curzip) == 2:
            return '000' + curzip
        elif len(curzip) == 3:
            return '00' + curzip
        elif len(curzip) == 4:
            return '0' + curzip
        else:
            return curzip
    except:
        return None

fio_21['POSTAL'] = fio_21.apply(convtostr, axis=1)
```

```python
fio_20['POSTAL'] = fio_20.apply(convtostr, axis=1)
fio_19['POSTAL'] = fio_19.apply(convtostr, axis=1)
fio_18['POSTAL'] = fio_18['zip']
```

```python
fio_21 = fio_21[fio_21['POSTAL'].notna()]
fio_20 = fio_20[fio_20['POSTAL'].notna()]
fio_19 = fio_19[fio_19['POSTAL'].notna()]
fio_18 = fio_18[fio_18['POSTAL'].notna()]
```

```python
import geopy
```

```python
def get_latlong(geolocator, zipcode):
    return geolocator.geocode({"postalcode": zipcode})

geolocator = geopy.Nominatim(user_agent='test')
```

```python
def get_lat_long(row):
    try:
        val = get_latlong(geolocator, row['POSTAL'])
        ret_lat = val.raw['lat']
        ret_long = val.raw['lon']
        return ret_lat, ret_long
    except:
        return None, None
```

```python
zipdict = {}
```

```python
some_data = fio_21
lats_2021 = []
lons_2021 = []
for index, row in some_data.iterrows():
    curzip = row['POSTAL']
    if curzip in zipdict:
        curlat,curlon = zipdict[curzip]
        lats_2021.append(curlat)
        lons_2021.append(curlon)
    else:
        curlat, curlon = get_lat_long(row)
        zipdict[curzip] = curlat, curlon
        lats_2021.append(curlat)
        lons_2021.append(curlon)
    if len(lats_2021) % 100 == 0:
        print(len(lats_2021))
```
