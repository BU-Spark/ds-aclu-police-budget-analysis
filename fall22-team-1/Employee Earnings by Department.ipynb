```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib import font_manager

font_dir = ['/Users/henry/Downloads/GTAmerica Font']
for font in font_manager.findSystemFonts(font_dir):
    font_manager.fontManager.addfont(font)
    
plt.rcParams['font.family'] = 'GT America'
```

```python
# 2021
x = pd.read_csv('2021earnings.csv')
x['TOTAL_GROSS'] = x['TOTAL_GROSS'].str.replace(',','')
x['TOTAL_GROSS'] = x['TOTAL_GROSS'].replace(np.nan, 0).astype(float)

xt = x[['DEPARTMENT_NAME', 'TOTAL_GROSS']]
xtd = xt.groupby(['DEPARTMENT_NAME'], as_index=False).sum(numeric_only=True).sort_values(by=['DEPARTMENT_NAME'])
dept_count = xt.DEPARTMENT_NAME.value_counts().sort_index().values
xtd['Average Gross'] = xtd['TOTAL_GROSS'] / dept_count

x10 = xtd.nlargest(10, 'Average Gross')

fig, ax = plt.subplots(figsize=(10,4))

plt.barh(x10['DEPARTMENT_NAME'], x10['Average Gross'], align='center', color='#130F54')
ax.invert_yaxis()
ax.get_children()[3].set_color('r')
ax.set_xlabel('Earnings in Dollars')
ax.set_title('Top 10 Average Earnings by Department in 2021')

plt.savefig('earnings2021.png', bbox_inches='tight')
plt.show()
```

```python
# 2020
x = pd.read_csv('2020earnings.csv')
x.rename(columns = {'TOTAL EARNINGS':'TOTAL_GROSS'}, inplace = True)
x['TOTAL_GROSS'] = x['TOTAL_GROSS'].str.replace(',','')
x['TOTAL_GROSS'] = x['TOTAL_GROSS'].str.replace('$','')
x['TOTAL_GROSS'] = x['TOTAL_GROSS'].replace(np.nan, 0).astype(float)

xt = x[['DEPARTMENT_NAME', 'TOTAL_GROSS']]
xtd = xt.groupby(['DEPARTMENT_NAME'], as_index=False).sum(numeric_only=True).sort_values(by=['DEPARTMENT_NAME'])
dept_count = xt.DEPARTMENT_NAME.value_counts().sort_index().values
xtd['Average Gross'] = xtd['TOTAL_GROSS'] / dept_count

x10 = xtd.nlargest(10, 'Average Gross')

fig, ax = plt.subplots(figsize=(10,4))

plt.barh(x10['DEPARTMENT_NAME'], x10['Average Gross'], align='center', color='#130F54')
ax.invert_yaxis()
ax.get_children()[2].set_color('r')
ax.set_xlabel('Earnings in Dollars')
ax.set_title('Top 10 Average Earnings by Department in 2020')

plt.savefig('earnings2020.png', bbox_inches='tight')
plt.show()
```
