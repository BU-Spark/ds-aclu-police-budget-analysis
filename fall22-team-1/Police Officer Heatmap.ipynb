```python
def convtostr(row):
    try:
        curzip = str(row['POSTAL'])
        if len(curzip) == 0:
            return '00000'
        elif len(curzip) == 1:
            return '0000' + curzip
        elif len(curzip) == 2:
            return '000' + curzip
        elif len(curzip) == 3:
            return '00' + curzip
        elif len(curzip) == 4:
            return '0' + curzip
        else:
            return curzip
    except:
        return None

all_officers_2019['POSTALSTR'] = all_officers_2019.apply(convtostr, axis=1)
```

```python
import geopy
```

```python
def get_latlong(geolocator, zipcode):
    return geolocator.geocode({"postalcode": zipcode})

geolocator = geopy.Nominatim(user_agent='test')
get_latlong(geolocator, '10117').raw['lon']
```

```python
def get_lat_long(row):
    try:
        val = get_latlong(geolocator, row['POSTAL'])
        ret_lat = val.raw['lat']
        ret_long = val.raw['lon']
        return ret_lat, ret_long
    except:
        return None, None
    
def get_lat_long_2019(row):
    try:
        val = get_latlong(geolocator, row['POSTALSTR'])
        ret_lat = val.raw['lat']
        ret_long = val.raw['lon']
        return ret_lat, ret_long
    except:
        return None, None
```

```python
%%time
some_data = all_officers_2021[:100]
zipdict = {}
lat_lons = []
for index, row in some_data.iterrows():
    curzip = row['POSTAL']
    if curzip in zipdict:
        lat_lons.append(zipdict[curzip])
    else:
        curlat, curlon = get_lat_long(row)
        zipdict[curzip] = curlat, curlon
        lat_lons.append((curlat, curlon))
```
